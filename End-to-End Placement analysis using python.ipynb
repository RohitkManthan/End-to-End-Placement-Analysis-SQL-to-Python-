### Table of package and percentage

```python
query = "SELECT * FROM percentage;" 
df = pd.read_sql(query, conn)
df
```

```python
query1 = "SELECT * FROM package;" 
df1 = pd.read_sql(query1, conn)
df1
```

# Create separate dataframes for Experienced and fresher students

```python
merged_df = pd.merge(df, df1, on='Roll_No')
fresher_df=merged_df[merged_df['workex']==False]
fresher_df
```

```python
experienced_df=merged_df[merged_df['workex']==True]
experienced_df
```

# Find out 3 fresher students who got maximum package.

```python
merged_df = pd.merge(df, df1, on='Roll_No')
fresher_students = merged_df[merged_df['workex'] == False]
top_3_fresher_packages = fresher_students.nlargest(3, 'salary')

print("Top 3 Fresher Students with the Highest Salary:")
top_3_fresher_packages



```

# What is average percentage in etest of experienced students?

```python
average_percentage=experienced_df['etest_p'].mean().round(2)
average_percentage
```

 # Calculate the difference in mba and degree percentage of students

```python
merged_df = pd.merge(df, df1, on='Roll_No')
merged_df['diff'] = merged_df['mba_p'] - merged_df['degree_p']
merged_df[['Roll_No', 'mba_p', 'degree_p', 'diff']]
```
